#### making function to define contract length

```python
def extract_contact_info(x):
    if "Free" in x or "On Loan" in x :
        start_date=np.nan
        end_date=np.nan
        contract_length=0
    else:
        year=x.split("~")
        start_date=year[0]
        end_date=year[1]
        contract_length=int(end_date)-int(start_date)
    return start_date,end_date,contract_length

new=pd.DataFrame(df["Contract"].apply(lambda x:pd.Series(extract_contact_info(x))))
df["start_date"]=new[0]
df["end_date"]=new[1]
df["contract_length"]=new[2]
```

```python
df[["Contract","start_date","end_date","contract_length"]].sample(10)
```

## Contract Status 
### Free, onloan, contract

```python
def status(x):
    if "Free" in x:
        return "Free"
    elif "On Loan" in x:
        return "On Loan"
    else:
       return "Contract"

df["Status"]=df["Contract"].apply(lambda x:status(x))
```

```python
df[["Contract","start_date","end_date","contract_length","Status",]].sample(20)
```

## Height
### in this column the value are strings and two types of value exist 

```python
df["Height"].unique()
```

```python
def convert_height(x):
    if 'cm' in x:
        return int(x.strip("cm"))
    else:
        fit,inches=x.split("'")
        total_inches=int(fit)*12+int(inches.strip('"')) ###converting fit into inches
        return (np.round(total_inches*2.54)) ###converting inches into cm

df["Height(cm)"]=df["Height"].apply(lambda x:convert_height(x))
        
```

```python
df[["Height","Height(cm)"]].sample(20)
```

## Weight

```python
df["Weight"].unique()
```

```python
def convert_weight(x):
    if 'kg' in x:
        return int(x.strip("kg"))
    else:
        lbs=x.strip("lbs")
        return (np.round(int(lbs)*2.205)) ###converting inches into cm

df["weight(kg)"]=df["Weight"].apply(lambda x:convert_weight(x))
```

## Value 
### Value paid to player  which are in million and thousand 

```python
df["Value"].unique()
```

```python
def value_changed(x):
    x=x.strip("€")
    if "M" in x:
        v=x.strip("M")
        cv=float(v)
        original_value=cv*1000000
    else:
        v=x.strip("K")
        cv=float(v)
        original_value=cv*1000

    return original_value

df["value"]=df["Value"].apply(lambda x:value_changed(x))
df[["Value","value"]]
        
```
